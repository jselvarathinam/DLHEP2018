### Exercise 1.1
Using random, write a function GenerateData(N, mymin, mymax), that returns a python list containing N random numbers between specified minimum and maximum value. Note that you may want to quickly work out on paper how to turn numbers between 0 and 1 to between other values. 

```python
# Skeleton
import random
    
def GenerateData(N,min,max):
    N = N + 1
    w = 1
    out = []
    while w < N:
        x=random.random()
        x = int(100 * x)
        if x < min:
            while x < min:
                x = x * 2
        if x > max:
            while x > max:
                 x = x / 3   
        out.append(x)
        w = w+1
    return out

Data=GenerateData(1000,-10,10)
print "Data Type:", type(Data)
print "Data Length:", len(Data)
if len(Data)>0: 
    print "Type of Data Contents:", type(Data[0])
    print "Data Minimum:", min(Data)
    print "Data Maximum:", max(Data)
```

### Exercise 1.2

Write a function that computes the mean of values in a list.

```python
# Skeleton

def mean(Data):
    m = 0
    i = 0
    g = len(Data)
    while g > i:
        m = m + Data[i]
        i = i + 1
    m = m/g
    return m



Data = [1,2,3,4,5,6,7,8,9]
print ("Mean of Data:", mean(Data))

```

### Exercise 1.4

The inrange(mymin,mymax) function below returns a function that tests if it's input is between the specified values. Use this function, in conjunction to your solution to 1.3, to demonstrate that your data is "flat". Hint: pick several sub-ranges and show that the number of data point divided by the size of the range is roughly constant. 

```python
import random

#First I will generate 1000 random numbers between -10 and 10 and store them in a list called "Data"
def GenerateData(N,min,max):
    N = N + 1
    w = 1
    out = []
    while w < N:
        x=random.random()
        x = int(100 * x)
        if x < min:
            while x < min:
                x = x * 2
        if x > max:
            while x > max:
                 x = x / 3   
        out.append(x)
        w = w+1
    return out
Data = GenerateData(1000,-10,10)

'''Now I create a function called "AnalyzeDataArray" that given a start and end index for a list of numbers will return 
a flaness index for the data between the start and end index. The flatness index is found by dividing the number of 
data points between the start and stop index by the range of the data sample.'''
def AnalyzeDataArray(start_index, end_index):
    data_subsection = Data[start_index:end_index] 
    sample_length = end_index - start_index
    range_of_sample = max(data_subsection) - min(data_subsection)
    flatness_index = sample_length/range_of_sample
    return flatness_index

#I now will calculate the flatness index for every 100 data points in my 1000 number list to demonstrate that the data is flat 
print(AnalyzeDataArray(0,100))
print(AnalyzeDataArray(100,200))
print(AnalyzeDataArray(200,300))
print(AnalyzeDataArray(300,400))
print(AnalyzeDataArray(400,500))
print(AnalyzeDataArray(600,700))
print(AnalyzeDataArray(700,800))
print(AnalyzeDataArray(800,900))
print(AnalyzeDataArray(900,999))
```

### Exercise 1.6

Write a new function called GenerateDataFromFunction(N,mymin,mymax,myfunc), that instead of generating a flat distribution, generates a distribution with functional form coded in myfunc. Note that myfunc will always be > 0.  

Use your function to generate 1000 numbers that are Gaussian distributed, using the Gaussian function below. Confirm the mean of the data is close to mean you specify when building the Gaussian. 

Hint: A simple, but slow, solution is to a draw random number test_x within the specified range and another number p between the min and max of the function (which you will have to determine). If p<=function(test_x), then place test_x on the output. If not, repeat the process, drawing two new numbers. Repeat until you have the specified number of generated numbers, N. For this problem, it's OK to determine the min and max by numerically sampling the function.  



```python
import math
import random

#This function will allow me to generate any amount of random numbers between a min and a max that I tell it
def GenerateData(N,min,max):
    N = N + 1
    w = 1
    out = []
    while w < N:
        x=random.random()
        x = int(100 * x)
        if x < min:
            while x < min:
                x = x * 2
        if x > max:
            while x > max:
                 x = x / 3   
        out.append(x)
        w = w+1
    return out

#This is the gaussian function
def gaussian(mean, sigma):
    def f(x):
        return math.exp(((x-mean)**2)/(2*sigma**2))/math.sqrt(math.pi*sigma)
    return f

#This function generates the 1000 numbers in a gaussian distribution 
def GenerateDataFromFunction(N,mymin,mymax,myfunc):
    i = 0
    out = []
    while N > i:
        x = GenerateData(1,-4,4)
        x = int(x[0])
        p = GenerateData(1,mymin,mymax)
        p = int(p[0])
        if p <= myfunc(x):
            out.append(myfunc(x))
            i = i + 1    
    return out

#Now I will solve and print the 1000 numbers that are gaussian distributed
solution = (GenerateDataFromFunction(1000,0,0.5,gaussian(0,4)))
print (solution)

#Now I will print the mean value of the gaussian distribution
print(sum(solution)/len(solution))
```
